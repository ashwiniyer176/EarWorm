```python
df.drop_duplicates(subset=["track","artist"],inplace=True)
df.dropna(inplace=True)
```

```python
def formatName(songName):
    songName=songName.split('-')
    print(songName[0].lower())
    return songName[0].lower()
```

```python
df.dropna(inplace=True)
```

```python
df.track=df['track'].apply(formatName)
df.artist=df.artist.apply(formatName)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df,test_df=train_test_split(df,test_size=0.25)
test_df,val_df=train_test_split(test_df,test_size=0.5)
```

```python
train_df.to_csv("./input/train.csv",index=False)
test_df.to_csv("./input/test.csv",index=False)
val_df.to_csv("./input/validation.csv",index=False)
```
