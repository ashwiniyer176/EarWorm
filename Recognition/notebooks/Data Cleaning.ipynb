```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../data/Song_Data_MFCC.csv")
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df.drop(columns=["filepath"],axis=1, inplace=True)
```

```python
X=df.drop("genre",axis=1)
y=df["genre"]
```

```python
scaler=MinMaxScaler()
scaled_features=scaler.fit_transform(X)
scaled_df=pd.DataFrame(scaled_features,columns=X.columns)
```

```python
scaled_df["genre"]=y
```

```python
scaled_df
```

```python
import os
replace_encoding={genre:idx for idx, genre in enumerate(os.listdir("../data/genres_original/"))}
```

```python
scaled_df["genre"].replace(replace_encoding,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test=train_test_split(scaled_df,test_size=0.25,random_state=69)
test,val=train_test_split(test,test_size=0.5,random_state=69)
```

```python
train=pd.concat([train,val])
```

```python
train.to_csv("./input/train.csv",index=False)
test.to_csv("./input/test.csv",index=False)
val.to_csv("./input/validation.csv",index=False)
```
